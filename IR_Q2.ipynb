```python
# Function to split number and string in a filename
def extract_number(filename):
    return int(''.join(filter(str.isdigit, filename)))

# Creating the unigram index
index = {}
for file in txt_files:
    with open(file, 'r') as f:
        content = f.read()
        words = content.split()
        file_id = extract_number(file)
        for word in words:
            if word.isalpha():

              if word not in index:
                  index[word] = []
              if file_id not in index[word]:
                  index[word].append(file_id)

```

```python
import os
mapping = {}
# Loop over files in the directory to create document mapping
for i, filename in enumerate(makeitlist, 1):
    mapping[i] = filename
```

```python
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Remove punctuations
    tokens = [token for token in tokens if token not in string.punctuation]
    # Remove blank space tokens
    tokens = [token for token in tokens if token.strip()]

    return tokens
```

```python
uni_set = {value for value in mapping.keys()}
```

```python
def AND(set1, set2):
    set_result = set1.intersection(set2)
    return set_result


def AND_NOT(set1, set2):
    set_result = set1.difference(set2)
    return set_result


def OR(set1, set2):
    set_result = set1.union(set2)
    return set_result


def OR_NOT(set1, set2):
    set_result = set1.union(set1, uni_set - set2)
    return set_result
```

```python
import re
import nltk


def process_query(query, operators, postings, universal_set):

    docs = []

    for i, op in enumerate(operators):
        op = op.strip()
        if i == 0:
            if query[i] in postings:
                word1_docs = postings[query[i]]
            else:
                word1_docs = []
            if query[i + 1] in postings:
                word2_docs = postings[query[i + 1]]
            else:
                word2_docs = []

            if op == "AND":
                docs= AND(set(word1_docs.copy()), set(word2_docs.copy()))
            elif op == "OR":
                docs= OR(set(word1_docs.copy()), set(word2_docs.copy()))
            elif op == "AND NOT":
                docs= AND_NOT(set(word1_docs.copy()), set(word2_docs.copy()))
            elif op == "OR NOT":
                docs= OR_NOT(set(word1_docs.copy()), set(word2_docs.copy()))
            else:
                print("Invalid Query!!!")
                return None, None
        else:
            if query[i + 1] in postings:
                word2_docs = postings[query[i + 1]]
            else:
                word2_docs = []
            if op == "AND":
                docs = AND(set(docs.copy()), set(word2_docs.copy()))
            elif op == "OR":
                docs= OR(set(docs.copy()), set(word2_docs.copy()))
            elif op == "AND NOT":
                docs= AND_NOT(set(docs.copy()), set(word2_docs.copy()))
            elif op == "OR NOT":
                docs= OR_NOT(set(docs.copy()), set(word2_docs.copy()))
            else:
                print("Invalid Query!!!")

                return None, None

    return docs

```
