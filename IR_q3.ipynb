```python
import os
import pickle
import re

# Function to preprocess text
def preprocess_text(text):
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens

# Function to create positional index
def create_positional_index(dataset_folder):
    positional_index = {}
    for filename in os.listdir(dataset_folder):
        filepath = os.path.join(dataset_folder, filename)
        if os.path.isfile(filepath):
            with open(filepath, 'r', encoding='latin-1') as file:  # Specify encoding
                text = file.read()
                tokens = preprocess_text(text)
                for position, token in enumerate(tokens):
                    if token not in positional_index:
                        positional_index[token] = {}
                    if filename not in positional_index[token]:
                        positional_index[token][filename] = []
                    positional_index[token][filename].append(position)
    return positional_index



# Save positional index to a file using pickle
def save_positional_index(positional_index, filepath):
    with open(filepath, 'wb') as f:
        pickle.dump(positional_index, f)

# Load positional index from a file using pickle
def load_positional_index(filepath):
    with open(filepath, 'rb') as f:
        positional_index = pickle.load(f)
    return positional_index


dataset_folder = '/content/drive/MyDrive/mod_file'

positional_index = create_positional_index(dataset_folder)
output_filepath = '/content/drive/MyDrive/positional_index.pkl'
save_positional_index(positional_index, output_filepath)

loaded_pos_index = load_positional_index(output_filepath)

print("Positional index created, saved, and loaded successfully.")
```

```python

def preprocess_text(text):
    tokens = re.findall(r'\b\w+\b', text.lower())
    return ' '.join(tokens)

def process_phrase_query(query_tokens, positional_index):
    results = set()
    for doc_id in positional_index.get(query_tokens[0], {}):
        positions = positional_index[query_tokens[0]][doc_id]
        for pos in positions:
            match = True
            for i in range(1, len(query_tokens)):
                next_token = query_tokens[i]
                next_positions = positional_index.get(next_token, {}).get(doc_id, [])
                if not next_positions:
                    match = False
                    break
                if pos + i not in next_positions:
                    match = False
                    break
            if match:
                results.add(doc_id)
                break
    return list(results)

# Function to process queries
def process_queries(query_list, positional_index):
    results = []
    for query in query_list:
        preprocessed_query = preprocess_text(query)
        query_tokens = preprocessed_query.split()
        query_results = process_phrase_query(query_tokens, positional_index)
        results.append(query_results)
    return results

# Input Format:
num_queries = int(input("Enter the number of queries: "))
query_list = []
for i in range(num_queries):
    query_list.append(input("Enter phrase query: "))

# Process queries
query_results = process_queries(query_list, loaded_pos_index)

# Output Format:
for i, query_result in enumerate(query_results):
    print(f"Number of documents retrieved for query {i+1} using positional index: {len(query_result)}")
    print(f"Names of documents retrieved for query {i+1} using positional index: {', '.join(query_result)}")
```
